```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'ppo_all_trajectories_test.csv'
df = pd.read_csv(file_path)

# Get 1/5 (20th percentile) and 4/5 (80th percentile) values across rows for each timestep (i.e., column-wise)
percentile_20 = df.quantile(0.2)
percentile_80 = df.quantile(0.8)

# Create date range for the x-axis
date_range = pd.date_range(start='2020-01-01', end='2023-01-01', periods=len(df.columns))

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(date_range, percentile_20, label='20th Percentile (1/5)', linestyle='--')
plt.plot(date_range, percentile_80, label='80th Percentile (4/5)', linestyle='-')
plt.xlabel('Date')
plt.ylabel('PF Value')
plt.title('1/5 and 4/5 Percentile of PF Values Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
